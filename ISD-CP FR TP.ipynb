```python
Total_Transfer = ISD2.groupby(['FROM_ATTRIBUTE']).agg({'Percent_Transfer': ['sum']})
Total_Transfer
```

```python
Total_Transfer.columns = [''.join(col).strip() for col in Total_Transfer.columns.values]
Total_Transfer
```

```python
Total_Transfer["max_value_column"] = Total_Transfer["Percent_Transfersum"].max()
Total_Transfer
```

```python
Total_Transfer["MAX SCALE DOWN FACTOR"] = Total_Transfer["max_value_column"] / 0.4
Total_Transfer
```

```python
ISD3 = pd.merge(ISD2, Total_Transfer, left_on='FROM_ATTRIBUTE', right_on='FROM_ATTRIBUTE', how='left')
ISD3
```

```python
ISD3["Scaled % DLR Transfer"] =  ISD3['Percent_Transfer'] / ISD3["MAX SCALE DOWN FACTOR"]
ISD3
```

```python
ISD3.loc[ISD3['max_value_column'] >= 0.4, 'Percent Transfer'] = ISD3['Scaled % DLR Transfer'] 
ISD3.loc[ISD3['max_value_column'] <= 0.4, 'Percent Transfer'] = ISD3['Percent_Transfer'] 

ISD3

#Percent Transfer without Underscore is Variable Name for the Export
```
