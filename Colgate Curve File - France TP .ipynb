```python
Raw_Long = Raw.set_index([('Info',  'Retailer'),
            ('Info',   'LDESC'),
            ('Info',   'ITEM'),


            ]).stack(dropna=False).reset_index().rename(columns = {'level_3': 'Date'})
Raw_Long

```

```python
dataTypeSeries = Raw_Long.dtypes
print(dataTypeSeries)
```

```python
Raw_Long.columns = [''.join(col).strip() for col in Raw_Long.columns.values]
Raw_Long
```

```python
Final = Raw_Long.rename(columns={'InfoRetailer':'Retailer','InfoITEM':'UPC_DESC','InfoLDESC':'UPC_SHORT_DESC', 

                                 'Date':'WEEK', 
                                   'Non Promo Unit':'UNITS', 'Non Promo Value':'DOLLAR', 'DV DIFFUSION [4S]':'ACV'})

Final
```

```python
Final2 = Final.groupby(['Retailer' , 'UPC_DESC' , 'WEEK']).agg({ 'UNITS':['sum'] , 'DOLLAR': ['sum'] , 'ACV':['mean']})
Final2
```

```python
Final2.columns = [''.join(col).strip() for col in Final2.columns.values]
Final2
```

```python
Final2.reset_index(inplace=True)
Final2
```

```python
Final3 = Final2.rename(columns={'UNITSsum':'UNITS', 'DOLLARsum':'DOLLAR', 'ACVmean':'ACV'})
Final3
```

```python
Final3.fillna(0, inplace = True)
Final3
```

```python
Final4 = Final3.loc[(Final3['UNITS'] != 0) 
& (Final3['DOLLAR'] != 0 )
& (Final3['ACV'] != 0 )  

 ]

Final4

#Final2 = Final.loc[(Final['UNITS'] != 0) 
#| (Final['SALES (EUR)'] != 0 )
#| (Final['ACV'] != 0 )  

# ]

Final4
```
